# US Arrests

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from scipy.stats import zscore
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
# Load the dataset.
df = pd.read_csv('UsArrests.csv')
df.head()
```

```python
# Are there any missing data?
df.isnull().sum()
```

```python
# That's good. But the 'city' column is clearly showing states, not cities.
df.City.unique()
```

```python
# So let's rename that.
df.rename(columns = {'City': 'State'}, inplace = True)
df.head()
```

```python
# What are our data types?
df.info()
```

```python
# Let's make sure the continuous variables are all in the same data type before we go any further.
df['Assault'] = df['Assault'].apply(float)
df['UrbanPop'] = df['UrbanPop'].apply(float)
df.info()
```

```python
# Let's get a sense of the correlation between our variables with a heatmap.
correlation = df.iloc[:, 1:].corr()
mask = np.triu(np.ones_like(correlation))

sns.heatmap(correlation, annot = True, mask = mask)
plt.show()
```

Urban population doesn't seem particularly correlated with anything else, but the other variables all seem quite well correlated with each other, which makes this a candidate for dimensionality reduction. But how many dimensions should we reduce to?

```python
# At the moment, the crime statistics are given as raw numbers per 100,000 population, and the urban population is a percentage.
# So, they're on different scales.
# So, we need to standardise the scores for our continuous variables.
for column in df.columns[1:5]:
    df[f'{column}Z'] = zscore(df[column])

df.head()
```

```python
# Now we carry out a PCA, using the standardised scores.
# At first, we keep all components.
exploratory_pca = PCA()
pd.DataFrame(exploratory_pca.fit_transform(df.iloc[:, 5:])).head()
```

```python
# We make a scree plot to see how much variance each component is explaining.
plt.plot(range(1, len(exploratory_pca.explained_variance_ratio_) + 1), 
         (np.cumsum(exploratory_pca.explained_variance_ratio_)) * 100)
plt.xlabel('Components')
plt.xticks(range(1, len(exploratory_pca.explained_variance_ratio_) + 1))
plt.ylabel('Explained variance (%)')
plt.title('Cumulative explained variance')
plt.show()
```

We get to over 85% explained variance with two components, and 95% with three. In what follows, I will proceed using two components, because

1. That's a substantial amount of variance already explained, and comfortably over the 80% threshold.
1. Two components are simpler to visualise.

In order to have a sense of what those components are encoding, we plot their correlation with the original four features.

```python
df.columns[5:]
```

```python
# Create a data frame that represents what we're looking at transparently.
pca_df = pd.DataFrame(exploratory_pca.fit_transform(df.iloc[:, 5:]), index = df.State).iloc[:, :2]
pca_df.rename(columns = {0: '1stPC', 1: '2ndPC'}, inplace = True)

for column in df.columns[5:]:
    pca_df[column] = df[column].values

pca_df.head()
```

```python
# Plot the correlation heat map.
pca_correlation = pca_df.corr()
pca_mask = np.triu(np.ones_like(pca_correlation))

sns.heatmap(pca_correlation, annot = True, mask = pca_mask)
plt.show()
```

Roughly speaking, then, the first principle component represents high violent crime statistics, and the second principle component represents a low urban population percentage.

Now we apply some clustering algorithms to establish what a sensible number of clusters is, and what they are.

### K-means clustering

```python
# We use the elbow method to find the right number of clusters.
# We'll try from 2 to 15 clusters, inclusive.
X = pca_df[['1stPC', '2ndPC']].values

elbow_xs = []
elbow_ys = []

for k in range(2, 16):
    km_model = KMeans(n_clusters = k, random_state = 0, n_init = 'auto')
    km_model.fit(X)
    inertia = km_model.inertia_
    elbow_xs.append(k)
    elbow_ys.append(inertia)
    
plt.plot(elbow_xs, elbow_ys, marker = 'o')
plt.xlabel('K')
plt.ylabel('Inertia')
plt.show()
```

The elbow comes at k = 4, so that's the number of clusters we will go on with.

#### Fitting, visualising and assessing the model

```python
# Fitting the model.
km_model = KMeans(n_clusters = 4, random_state = 0, n_init = 'auto')
km_y_pred = km_model.fit_predict(X)

# Recording the predicted categories.
pca_df['KmeansPred'] = km_y_pred
pca_df.head()
```

```python
# Visualising the predictions.
plt.figure(figsize = (10, 6))
plt.scatter(pca_df['1stPC'], pca_df['2ndPC'], c = pca_df['KmeansPred'])

for row in range(len(pca_df)):
    plt.annotate(pca_df.index[row], xy = (pca_df.iloc[row, 0], pca_df.iloc[row, 1]))

plt.xlabel("First principle component")
plt.ylabel("Second principle component")
plt.title('K-Means Clustering')
plt.show()
```

To a first impression, the clusters seem to be quite well-defined, although one may wonder about the classification of Arkansas, Kentucky and Montana. We can evaluate the clustering more objectively with a silhouette score.

```python
print(f"K-means silhouette score: {silhouette_score(X, km_y_pred)}")
```

The silhouette score is OK but not fantastic, which adds to the sense that perhaps some data points could be better categorised. One hypothesis is that we are being let down here by the assumption of spherical clusters. We will revisit this idea when we apply a different clustering algorithm later on.

In order to get a sense of what our clusters represent, we can group the dataset by cluster and see if the clusters have any defining characteristics in terms of the original statistics.

```python
k_means_cats = pca_df.groupby(by = 'KmeansPred').mean()
k_means_cats
```

For example, the green cluster scores high on the first principle component, which we said meant high violent crime statistics, and high on the second component, which we said meant low urban population.

### Agglomerative hierarchical clustering

We will apply another method (in addition to the elbow method above) to evaluate how many clusters our data should have.

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (15, 6))

dendrogram(linkage(X, method = 'single', metric = 'euclidean'), ax = ax1, labels = pca_df.index)
ax1.set_title("Single linkage")
dendrogram(linkage(X, method = 'average', metric = 'euclidean'), ax = ax2, labels = pca_df.index)
ax2.set_title("Average linkage")
dendrogram(linkage(X, method = 'complete', metric = 'euclidean'), ax = ax3, labels = pca_df.index)
ax3.set_title("Complete linkage")

plt.show()
```

It's clearest to tell visually from complete linkage where the distances between clusters start to get much larger: there is one orange line and two green lines with a size of around 2 or larger. If we cut those lines we get four clusters (as before), so that's what we will go with.

```python
# Fitting the model.
agglom_model = AgglomerativeClustering(n_clusters = 4, metric = 'euclidean', linkage = 'complete')
agglom_y_pred = agglom_model.fit_predict(X)

# Recording the predicted categories.
pca_df['AgglomPred'] = agglom_y_pred
pca_df.head()
```

We can already see some disagreement between the two models.

```python
# Visualising the predictions.
plt.figure(figsize = (10, 6))
plt.scatter(pca_df['1stPC'], pca_df['2ndPC'], c = pca_df['AgglomPred'])

for row in range(len(pca_df)):
    plt.annotate(pca_df.index[row], xy = (pca_df.iloc[row, 0], pca_df.iloc[row, 1]))

plt.xlabel("First principle component")
plt.ylabel("Second principle component")
plt.title('Agglomerative Hierarchical Clustering')
plt.show()
```

The first thing that strikes me from this clustering is that one cluster is noticeably larger than the others. Unlike K-means clustering, agglomerative hierarchical clustering does not assume that the prior probability for each cluster is the same, and this is the result. Relatedly, the three states that we wondered about above&mdash;Arkansas, Kentucky and Montana&mdash;are in this largest cluster.

Let's once again group the dataset by cluster and see if the clusters have any defining characteristics in terms of the original statistics. At the same time, let's compare this grouping to the one above based on k-means clustering.

```python
agglom_cats = pca_df.groupby(by = 'AgglomPred').mean()
agglom_cats
```

The four clusters seem to be tracking the same features as with the previous clustering. We can compare the two clustering with respect to how clearly they track those features.

```python
print(f'''
Low violent crime, low urban population.

K-Means:
{k_means_cats.iloc[0, 2:6]}

Agglomerative Hierarchical:
{agglom_cats.iloc[3, 2:6]}


Low violent crime, high urban population.

K-Means:
{k_means_cats.iloc[1, 2:6]}

Agglomerative Hierarchical:
{agglom_cats.iloc[0, 2:6]}


High violent crime, low urban population.

K-Means:
{k_means_cats.iloc[2, 2:6]}

Agglomerative Hierarchical:
{agglom_cats.iloc[2, 2:6]}


High violent crime, high urban population.

K-Means:
{k_means_cats.iloc[3, 2:6]}

Agglomerative Hierarchical:
{agglom_cats.iloc[1, 2:6]}
''')
```

In the vast majority of cases, the absolute value is higher on the agglomerative hierarchical clustering than on the k-means clustering. For example, for the low violent crime, low urban population cluster, the values are more strongly negative on the agglomerative hierarchical clustering. And for the high violent crime, high urban population cluster, the values are more strongly positive on the agglomerative hierarchical clustering. This would seem to indicate that the agglomerative hierarchical clustering is capturing these natural clusters better. We can check this by taking the silhouette score.

```python
print(f"Agglomerative hierarchical silhouette score: {silhouette_score(X, agglom_y_pred)}")
```

The silhouette score is actually not as good as the one for the k-means algorithm, but similar. Overall, we have to say that these data are really quite evenly spread out in general.
